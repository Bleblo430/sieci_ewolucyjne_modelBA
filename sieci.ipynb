```python
#stara wersja


M = 1 #liczba węzłów które dołączomy w jednej iteracji
K = 1 #liczba karawędzi dołączanego węzła 


N = 100000 #liczba iteracji
duplicates = True
preferencja = True
#punkt startowy
graph = [
    [1], #węzeł 0 połączony z węzłem 1 
    [0] #węzeł 1 połączony z węzłem 0
]

propability_table = [0,1]




for i in range(N): #dla każdej z N iteracji sieci sprawdzamy liczbę dostępnych węzłów do których będziemy tworzyć połączenia
    num_of_slots = len(graph)
    if i % 1000 == 0:
        print(i)
    if preferencja == False:
        for j in range(M): #tworzymy M nowych węzłów
            new_node = len(graph) #indexy są numerowane od 0 natomiast długość zwraca wartość od 1 więc len(graph) zwraca index nowego węzła 
            graph.append([]) #dodajemy nowy węzeł do graph
            K_safe = min(num_of_slots,K) #jeśli chcemy tworzyć więcej połączeń w niż istnieje węzłów targets się wywali więc jeśli K>num of slots, num of slots przyjmuje jego rolę 
            if duplicates == True:
                targets = np.random.choice(num_of_slots, K_safe, replace=True) #dla każdego nowego węzła losujemy K węzłów z już istniejących do których będą kreowane połączenia
            elif duplicates == False:
                targets = np.random.choice(num_of_slots, K_safe, replace=False)
            for slot in targets:
                graph[new_node].append(int(slot)) #targets zwraca tablice np.array więc warto to zrzutować na zmienne pythonowe 
                graph[slot].append(new_node)
                
    elif preferencja == True:
        #logika ta sama tylko zamiast losować targets z zakresu 0 do sym_of_slots gdzie każa ma to samo prawdopodobieństwo losujemy liczby z propability table    
        for j in range(M):
            new_node = len(graph)
            graph.append([])
            K_safe = min(num_of_slots,K)
            if duplicates == True:
                targets = np.random.choice(propability_table, K_safe, replace=True)
            elif duplicates == False:
                targets = []
                while len(targets) < K_safe:
                    slot = int(np.random.choice(propability_table))
                    if slot not in targets:
                        targets.append(slot)
            for slot in targets:
                slot = int(slot)
                graph[new_node].append(slot) 
                graph[slot].append(new_node)    
                propability_table.append(slot)
                propability_table.append(new_node)        



```
